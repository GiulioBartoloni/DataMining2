```python
import pandas as pd

loda = pd.read_csv('loda.csv')
lof = pd.read_csv('lof.csv')
eif = pd.read_csv('cmi_eif_scores.csv')
```

```python
loda = loda.loc[loda['is_outlier']==1]
lof = lof.loc[lof['outlier']==1]
eif = eif.loc[eif['is_outlier']==1]
```

```python
loda['Unnamed: 0']
```

```python
loda_m = loda[['Unnamed: 0']].assign(loda=1)
lof_m = lof[['Unnamed: 0']].assign(lof=1)
eif_m = eif[['Unnamed: 0']].assign(eif=1)

merged = (loda_m
    .merge(lof_m, on='Unnamed: 0', how='outer')
    .merge(eif_m, on='Unnamed: 0', how='outer')
    .fillna(0))

merged['count'] = merged[['loda', 'lof', 'eif']].sum(axis=1).astype(int)
result = merged[merged['count'] >= 1].reset_index(drop=True)
result
```
